```python
## ì¥ì†Œ ê²€ìƒ‰ í•¨ìˆ˜ findPlaces
def findPlace(shopName):
    # 1. ê²€ìƒ‰API ì£¼ì†Œë³€ìˆ˜
    url = f'https://dapi.kakao.com/v2/local/search/{encodeKeyword}.json?query={shopName}'
    # 2. header êµ¬ì„±
    header = { 'Authorization' : 'KakaoAK 24d1b70d32491da959aac9405ab101f1'}
    # 3. APIë¥¼ ì›¹ìœ¼ë¡œ ìš”ì²­, ê²°ê³¼ ë¦¬í„´
    places = requests.get(url, headers=header).json()['documents']
    #4. í•„ìš”ì •ë³´ë§Œ ì¶”ì¶œ
    place = places[0]
    realName=place['place_name']
    lat = place['y'] # ìœ„ë„(latitude)
    lng = place['x'] # ê²½ë„(longitude)
    phoneNum = place['phone']
    realAddress = place['address_name']

    return [shopName, realName, phoneNum, realAddress, lat, lng]
    
    
```

```python
## í…ŒìŠ¤íŠ¸
result = findPlace('1100ê³ ì§€')
result
```

```python
# Locations ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜ë³µë¬¸ìœ¼ë¡œ ì£¼ì†Œê°€ì ¸ì˜¤ëŠ” ì²˜ë¦¬
locationResults = []

for loc in tqdm(Locations):
    try:
        data = findPlace(loc)
        locationResults.append(data)
        time.sleep(0.3)
    except:
        pass
```

```python
## ì¤‘ë³µë°ì´í„° pivot
locationDataFinal = locationData.pivot_table(values='count', index=['ì‹¤ì œì´ë¦„','ì£¼ì†Œ','ìœ„ë„','ê²½ë„'], aggfunc='sum')
```

```python
# ìœ„ë„ ê²½ë„ê°€ objectì—ì„œ float64ë¡œ ë³€ê²½
locationDataFinal = locationDataFinal.astype({'ìœ„ë„': 'float64', 'ê²½ë„': 'float64'},)
```

##### ì§€ë„ ì‹œê°í™”

```python
locHanlaMt = [33.362050, 126.530087]
map = folium.Map(location = locHanlaMt, zoom_start = 11, tiles = 'OpenStreetMap')

for i in range(len(locationDataFinal)):
    name = locationDataFinal['ì‹¤ì œì´ë¦„'][i] # DF ì»¬ëŸ¼ì´ë¦„ í–‰ê°’ == i 
    count = int(locationDataFinal['count'][i])
    lat = float(locationDataFinal['ìœ„ë„'][i])
    lng = float(locationDataFinal['ê²½ë„'][i])
    folium.CircleMarker(location=[lat, lng], radius=count, color='red', popup=name).add_to(map) #
map
```

```python
map.save('./data/3_jeju_hotplaces.html')
```

```python
## ë§ˆì»¤ í´ëŸ¬ìŠ¤í„°
mapLocs = []
mapNames = []

for i in range(len(locationDataFinal)):
    data = locationDataFinal.iloc[i] # í–‰ì„ í•˜ë‚˜ì”© ì½ì–´ì˜´
    mapLocs.append((float(data['ìœ„ë„']), float(data['ê²½ë„'])))
    mapNames.append(data['ì‹¤ì œì´ë¦„'])
    
locHanlaMt = [33.362050, 126.530087]
map = folium.Map(location = locHanlaMt, zoom_start = 11, tiles = 'OpenStreetMap')

makerCluster = MarkerCluster(locations=mapLocs, popups=mapNames, name='ì œì£¼ í•«í”Œë ˆì´ìŠ¤', overlay=True, control=True)
makerCluster.add_to(map)
folium.LayerControl().add_to(map)

map
```
